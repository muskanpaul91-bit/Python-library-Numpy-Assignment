```python
Use NumPy arrays
```

## Sample Output


# Q10. Concatenation and Stacking

## Problem Statement
Perform concatenate and vstack on two arrays.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
print(np.concatenate([a,b]))
````

````python
print(np.vstack([a,b]))
````

````python
print(np.hstack([a,b]))
````
